```python
#matplotlib.pyplotメソッドとnumpyをインポートするコードを記述する。
import numpy as np
import matplotlib.pyplot as plt

#描画領域全体に、2rows×2cols個のサブプロットを生成する
fig, axes = plt.subplots(2, 3, tight_layout=True)
#折れ線グラフ
y0 = [1, 2, -5, 2]
axes[0,0].plot(y0)
axes[0,0].set_title("折れ線グラフ")

#sin関数
x = np.linspace(0, 10, 100)
y = 2 + 2 * np.sin(2 * x)
axes[0,1].plot(x, y)
axes[0,1].set_title("sin関数")

#ヒストグラム
x = np.random .normal(15, 5, 2000)
axes[0,2].hist(x)
axes[0,2].set_title("ヒストグラム")

#散布図
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 60)
y = 4 + np.random.normal(0, 2, len(x))
axes[1,0].scatter(x, y)
axes[1,0].set_title("散布図")

#グラフを表示
plt.show()
```
